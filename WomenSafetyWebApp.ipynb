```python
#dropping Helpline Number
#drop Cities
#drop Police Station
#drop Helpline


train = train.drop(['Helpline Number','Cities','Nearest Police Station'], axis = 1)
train
```

```python
#camera survelliance
# "YES" == 1
# "NO" == 0

def getcameraSurvelliance(str):
    if str == "YES":
        return 1
    elif str == "NO":
        return 0
    
```

```python
train['Camera Survelliance'] = train['Camera Survelliance'].apply(getcameraSurvelliance)
print(train['Camera Survelliance'])
```

```python
#district to numbers -->> 
# east - 1
# north -2
# north west - 3
# south - 4
# south west - 5
# west - 6

def getDistrictNumbers(str):
    if str == "East":
        return 1
    elif str == "North":
        return 2
    elif str == "North West":
        return 3
    elif str == "South":
        return 4
    elif str == "South West":
        return 5
    elif str == "West":
        return 6
 
```

```python
train['District'] = train['District'].apply(getDistrictNumbers)
print(train['District'])
```

```python
train_data = pd.DataFrame(train.dtypes)
train_data
```

```python
Xtrain, Ytrain = np.split(train, [-1], axis =1)
```

```python
Xtrain
```

```python
Ytrain
```

```python
scaler = StandardScaler()
Xtrain['Crowd'] = scaler.fit_transform(Xtrain['Crowd'].values.reshape(-1,1))
```

```python
#decision tree classifier
clf = DecisionTreeClassifier()
```

```python
clf.fit(Xtrain, Ytrain)
```
